```python
import tensorflow as tf
from tensorflow import keras
```

```python
from tensorflow.keras.backend import set_session
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
sess = tf.Session(config=config)
set_session(sess)
```

```python
def build_generator():
    noise = keras.layers.Input(shape=(100,))
    label_class = keras.layers.Input(shape=(1,))
    
    dense1 = keras.layers.Dense(128, activation='tanh')(noise)
    concat = keras.layers.Concatenate()([dense1, label_class])
    dense2 = keras.layers.Dense(64, activation='tanh')(dense1)
    
    out = keras.layers.Dense(29)(dense2) #output is more than 0, and more than 1
    
    return keras.models.Model([noise, label_class], out)
```

```python
def build_discriminator():
    data = keras.layers.Input(shape=(29,))
    label_class = keras.layers.Input(shape=(1,))
    
    inp = keras.layers.Concatenate()([data, label_class])
    
    dense1 = keras.layers.Dense(32, activation='relu')(inp)
    dense2 = keras.layers.Dense(16, activation='relu')(dense1)
    
    out = keras.layers.Dense(1, activation='sigmoid')(dense2)
    
    return keras.models.Model([data, label_class], out)
```

```python
discriminator_model = build_discriminator()
discriminator_model.compile(loss=['binary_crossentropy'], optimizer=keras.optimizers.RMSprop(0.001), metrics=['accuracy'])
```

```python
generator_model = build_generator()
```

```python
noise = keras.layers.Input(shape=(100,))
label_class = keras.layers.Input(shape=(1,))
data = generator_model([noise, label_class])

discriminator_model.trainable = False

valid = discriminator_model([data, label_class])

combined_model = keras.models.Model([noise, label_class], valid)
combined_model.compile(loss=['binary_crossentropy'], optimizer=keras.optimizers.RMSprop(0.01), metrics=['accuracy'])
```

```python
import numpy as np
```

```python
negative_label = np.zeros(shape=(492, 1))
positive_label = np.ones(shape=(492, 1))
```

```python
positive_data = positive.drop(['Class'], axis=1)
negative_data = negative.drop(['Class'], axis=1)
```

```python
import random
from sklearn.utils import shuffle
```

```python
epoch = 1
iteration = 0
random_noise = []
random_class = []
iteration=0

while(True):
    if(epoch%10==0 or epoch==1): print(epoch, end=' ')
    else: print('.', end=' ')
    
    if(epoch%500==0 or epoch==1):
        iteration=0
        negative_data = shuffle(negative_data)
        
    random_noise = np.random.random_sample((492, 100))
    random_class = np.random.randint(2, size=(492, 1))
    gen_data = generator_model.predict([random_noise, random_class])
    
    # Train real data
    _ = discriminator_model.fit(x=[positive_data, positive_label], y=positive_label, verbose=0, batch_size=64)
    _ = discriminator_model.fit(x=[negative_data[iteration*492:iteration*492+492], negative_label], y=positive_label, verbose=0, batch_size=64)
    # Train fake data
    d_history = discriminator_model.fit(x=[gen_data, random_class], y=negative_label, verbose=0, batch_size=32)
    
    # discriminator train 4 times
    # Train positive class
    _ = combined_model.fit(x=[random_noise, negative_label], y=positive_label, verbose=0, batch_size=32)
    # Train negative class
    g_history = combined_model.fit(x=[random_noise, positive_label], y=positive_label, verbose=0, batch_size=32)
    
    if(epoch%50==0):
        print()
        print("discriminator acc: \t", int(d_history.history['acc'][-1]*100))
        print("discriminator loss: \t", d_history.history['loss'][-1])
        print("generator acc: \t\t", int(g_history.history['acc'][-1]*100))
        print("generator loss: \t", g_history.history['loss'][-1])
    
    if(d_history.history['acc'][-1]>0.45 and d_history.history['acc'][-1]<0.55):
        break
    epoch+=1
    iteration+=1
```
